```python
import pandas as pd
import seaborn as sns # Veri setini yüklemek ve görselleştirme için
import matplotlib.pyplot as plt
```

```python
# ---------------------------------------------------------
# ADIM 1: VERİYİ YÜKLEME VE İLK BAKIŞ
# ---------------------------------------------------------
print("--- 1. VERİYİ YÜKLEME VE İLK BAKIŞ ---")
# Seaborn kütüphanesinin içindeki hazır Titanic veri setini yüklüyoruz.
df = sns.load_dataset('titanic')
```

```python
# Verinin boyutunu öğrenelim (Satır, Sütun)
print(f"Veri Seti Boyutu: {df.shape[0]} satır, {df.shape[1]} sütun.\n")
```

```python
# Verinin ilk 5 satırına bakalım
print("Verinin İlk 5 Satırı:")
print(df.head(), "\n")
```

```python
# Veri tipleri ve genel bilgi
print("Veri Seti Bilgisi (Info):")
df.info() 
print("\n")
```

```python
# ---------------------------------------------------------
# ADIM 2: İSTATİSTİKSEL ÖZET VE EKSİK VERİ KONTROLÜ
# ---------------------------------------------------------
print("--- 2. İSTATİSTİKSEL ÖZET VE EKSİK VERİLER ---")
```

```python
# Sayısal sütunların istatistiksel özeti (Ortalama, min, max, standart sapma)
print("İstatistiksel Özet:")
print(df.describe(), "\n")
```

```python
# Hangi sütunda kaç tane eksik (NaN) değer var?
print("Eksik Değerlerin Sayısı:")
print(df.isnull().sum(), "\n")
```

```python
# ---------------------------------------------------------
# ADIM 3: TEK DEĞİŞKENLİ ANALİZ (Univariate)
# ---------------------------------------------------------
print("--- 3. TEK DEĞİŞKENLİ ANALİZ ---")
```

```python
# Yolcuların cinsiyet dağılımı (Kategorik bir değişken analizi)
print("Cinsiyet Dağılımı:")
print(df['sex'].value_counts(), "\n")
```

```python
# Hayatta kalma oranları (0: Öldü, 1: Hayatta Kaldı)
print("Hayatta Kalma Durumu:")
print(df['survived'].value_counts(), "\n")
```

```python
# ---------------------------------------------------------
# ADIM 4: ÇOK DEĞİŞKENLİ ANALİZ (Bivariate/Multivariate)
# ---------------------------------------------------------
print("--- 4. ÇOK DEĞİŞKENLİ ANALİZ ---")
```

```python
# Cinsiyete göre hayatta kalma oranlarının ortalaması (Kadınlar vs Erkekler)
print("Cinsiyete Göre Hayatta Kalma Oranları:")
print(df.groupby('sex')['survived'].mean(), "\n")
```

```python
# Sınıflara (1., 2., 3. sınıf) göre bilet ücreti (fare) ortalaması
print("Yolcu Sınıfına Göre Ortalama Bilet Ücreti:")
print(df.groupby('class')['fare'].mean(), "\n")
```

```python
# Sadece sayısal sütunları seçip Korelasyon (İlişki) matrisine bakalım
numeric_df = df.select_dtypes(include=['float64', 'int64'])
print("Sayısal Değişkenler Arası Korelasyon:")
print(numeric_df.corr())
```

# Kodun Çıktılarından Çıkarılacak Bazı EDA Sonuçlar:
1.  Eksik Veriler: Kodu çalıştırdığınızda age (yaş) ve deck (güverte) sütunlarında çok fazla eksik veri olduğunu göreceksiniz. Bu bize, model kurmadan önce bu eksiklikleri doldurmamız veya deck sütununu komple silmemiz gerektiği ipucunu verir.

2.  Çok Değişkenli Analiz Çıkarımı: df.groupby('sex')['survived'].mean() kodunun çıktısı kadınların hayatta kalma oranının erkeklere göre çok daha yüksek olduğunu gösterecektir (Gemideki "Önce kadınlar ve çocuklar" kuralının veriyle ispatı).

3.   Korelasyon: fare (bilet ücreti) ile survived (hayatta kalma) arasında pozitif bir korelasyon görülür. Yani daha pahalı bilet alanların (1. sınıf yolcular) hayatta kalma ihtimali daha yüksek olmuştur.
