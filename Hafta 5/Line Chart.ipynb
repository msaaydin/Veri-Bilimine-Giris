### Line Chart

A line chart displays the evolution of one or several numeric variables. It is one of the most common chart type, so it can be build using any python viz library, like matplotlib, seaborn.

The flights dataset has 10 years of monthly airline passenger data

```python
import seaborn as sns

# Read Data
flights = sns.load_dataset("flights")
flights.head()
```

#### Matplotlib

Details: https://matplotlib.org/3.5.1/api/_as_gen/matplotlib.pyplot.plot.html

```python
# libraries
import matplotlib.pyplot as plt

# Get lines that 'month' value is equal to 'May'
filtered_data = flights[flights['month'] == 'May']
x = filtered_data["year"] # X-axis: Year
y = filtered_data["passengers"] # Y-axis: Number of passengers

# use the plot function - Matplotlib version
plt.xlabel("Year") # Set the label for the x-axis.
plt.ylabel("Number of Passengers") # Set the label for the y-axis.
plt.title("Year vs Number of Passengers") # Set the title
plt.plot(x, y) # Line Chart
plt.show()
```

#### Line Properties

**marker:** Representing marker types, Details: https://matplotlib.org/3.5.1/api/_as_gen/matplotlib.markers.MarkerStyle.html#matplotlib.markers.MarkerStyle

**color:** List of the named colors supported in matplotlib, Details: https://matplotlib.org/stable/gallery/color/named_colors.html

**linestyles:** Simple linestyles can be defined using the strings 'solid', 'dotted' etc., Details: https://matplotlib.org/stable/gallery/color/named_colors.html

```python
# libraries
import matplotlib.pyplot as plt

# Get lines that 'month' value is equal to 'May'
filtered_data = flights[flights['month'] == 'May']
x = filtered_data["year"] # X-axis: Year
y = filtered_data["passengers"] # Y-axis: Number of passengers

# use the plot function - Matplotlib version
plt.xlabel("Year") # Set the label for the x-axis.
plt.ylabel("Number of Passengers") # Set the label for the y-axis.
plt.title("Year vs Number of Passengers") # Set the title
plt.plot(x, y, marker = '^', color = 'r', linestyle='dashed') # Line Chart
plt.show()
```

#### Seaborn

```python
import seaborn as sns

# Read Data
flights = sns.load_dataset("flights")
flights.head()

# Get lines that 'month' value is equal to 'May'
filtered_data = flights[flights['month'] == 'May']

sns.lineplot(data=filtered_data, x="year", y="passengers")
```

#### Line Properties
Details: https://seaborn.pydata.org/generated/seaborn.lineplot.html

```python
import seaborn as sns

# Read Data
flights = sns.load_dataset("flights")
flights.head()

# Get lines that 'month' value is equal to 'May'
filtered_data = flights[flights['month'] == 'May']

sns.lineplot(data=filtered_data, x="year", y="passengers", marker='o', color = 'r', linestyle='dashed')
```

#### Multiple Line Plot

#### Matplotlib

```python
may_data = flights.query("month == 'May'")
july_data = flights.query("month == 'Jul'")

plt.plot(may_data['year'], may_data['passengers'], marker = 'o', color='b')
plt.plot(july_data['year'], july_data['passengers'], marker = 'o', color='r')
plt.legend(['May', 'July']) # Place a legend on the Axes.
plt.show()
```

#### Seaborn

```python
may_data = flights.query("month == 'May'")
july_data = flights.query("month == 'Jul'")

sns.lineplot(data=may_data, x='year', y='passengers', marker = 'o', color='b')
sns.lineplot(data=july_data, x='year', y='passengers', marker = 'o', color='r')
plt.legend(['May', 'July'])
```
