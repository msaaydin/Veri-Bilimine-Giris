### Barplot - Ranking

A barplot shows the relationship between a numeric and a categoric variable. Each entity of the categoric variable is represented as a bar. The size of the bar represents its numeric value.

```python
import matplotlib.pyplot as plt
import numpy as np

# Make a random dataset:
languages =['Python', 'SQL', 'Java', 'C++', 'JavaScript']
pos = np.arange(len(languages))
popularity = [56, 39, 34, 34, 29]

# Create bars
plt.bar(pos, popularity, align='center')

# Create names on the x-axis
plt.xticks(pos, languages)

plt.xlabel("Programming Languages")
plt.ylabel('% Popularity')

plt.title('Top 5 Languages for Math & Data \nby % popularity on Stack Overflow', alpha=0.8)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Make a random dataset:
languages =['Python', 'SQL', 'Java', 'C++', 'JavaScript']
pos = np.arange(len(languages))
popularity = [56, 39, 34, 34, 29]

# Create bars
bars = plt.bar(pos, popularity, align='center', color = ['red', 'blue', 'green', 'orange', 'purple'])

# Create names on the x-axis
plt.xticks(pos, languages)

plt.ylabel('% Popularity')

plt.title('Top 5 Languages for Math & Data \nby % popularity on Stack Overflow', alpha=0.8)

# direct label each bar with Y axis values
for bar in bars:
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() - 5, str(int(bar.get_height())) + '%', 
                 ha='center', color='w', fontsize=11)
    

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Create bars
barWidth = 0.9
bars1 = [65, 18, 17] 
bars2 = [56, 39, 5] 
bars3 = [45, 35, 20] 
bars4 = bars1 + bars2 + bars3

# The X position of bars
r1 = [1,5,9]
r2 = [2,6,8]
r3 = [3,4,7]
r4 = r1 + r2 + r3

# Make a random dataset:
languages =['Javascript', 'PHP', 'Java', 'Swift', 'Java', 'Objective-C', 'Python', 'C#', 'C++']

# Create barplot
bar1 = plt.bar(r1, bars1, width = barWidth, color = (0.3,0.1,0.4,0.6), label='Web Programming')
bar2 = plt.bar(r2, bars2, width = barWidth, color = (0.3,0.5,0.4,0.6), label='Mobile Programming')
bar3 = plt.bar(r3, bars3, width = barWidth, color = (0.3,0.9,0.4,0.6), label='Desktop Programming')
bars = [bar1, bar2, bar3]

# Create legend
plt.legend()

# Text below each barplot with a rotation at 90°
plt.xticks([r + barWidth for r in range(len(r4))], languages, rotation=90)

plt.ylabel('% Popularity')

# direct label each bar with Y axis values
for bar in bars:
    for b in bar:
        plt.text(b.get_x() + b.get_width()/2, b.get_height() + 2, str(int(b.get_height())) + '%', 
                     ha='center', color='b', fontsize=11)
    

plt.show()
```

#### Seaborn

```python
import seaborn as sns
sns.set_theme(style="whitegrid")
tips = sns.load_dataset("tips")
ax = sns.barplot(x="day", y="total_bill", data=tips)
```
