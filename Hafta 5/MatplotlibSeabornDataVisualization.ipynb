### Plotting from a script

If you are using Matplotlib from within a script, the function ``plt.show()`` is your friend.
``plt.show()`` starts an event loop, looks for all currently active figure objects, and opens one or more interactive windows that display your figure or figures.

So, for example, you may have a file called *myPlot.py* containing the following:

```python
# ------- file: myPlot.py ------
import matplotlib.pylot as plt
import numpy as np

x = np.linspace(0, 10, 100)

plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))

plt.show()
```

You can then run this script from the command-line prompt, which will result in a window opening with your figure displayed:

```
$ python myPlot.py
```

The ``plt.show()`` command does a lot under the hood, as it must interact with your system's interactive graphical backend.
The details of this operation can vary greatly from system to system and even installation to installation, but matplotlib does its best to hide all these details from you.

One thing to be aware of: the ``plt.show()`` command should be used *only once* per Python session, and is most often seen at the very end of the script.
Multiple ``show()`` commands can lead to unpredictable backend-dependent behavior, and should mostly be avoided.

### Matplotlib API Overview


Matplotlib has two APIs to work with. A MATLAB-style state-based interface and a more powerful object-oriented (OO) interface. 
The former MATLAB-style state-based interface is called **pyplot interface** and the latter is called **Object-Oriented** interface.


There is a third interface also called **pylab** interface. It merges pyplot (for plotting) and NumPy (for mathematical functions) together in an environment closer to MATLAB. This is considered bad practice nowadays. So, the use of **pylab** is strongly discouraged and hence, I will not discuss it any further.


### Pyplot API 


**Matplotlib.pyplot** provides a MATLAB-style, procedural, state-machine interface to the underlying object-oriented library in Matplotlib. **Pyplot** is a collection of command style functions that make Matplotlib work like MATLAB. Each pyplot function makes some change to a figure - e.g., creates a figure, creates a plotting area in a figure etc. 


**Matplotlib.pyplot** is stateful because the underlying engine keeps track of the current figure and plotting area information and plotting functions change that information. To make it clearer, we did not use any object references during our plotting we just issued a pyplot command, and the changes appeared in the figure.


We can get a reference to the current figure and axes using the following commands-


`plt.gcf ( )`   # get current figure

`plt.gca ( )`   # get current axes 

 
**Matplotlib.pyplot** is a collection of commands and functions that make Matplotlib behave like MATLAB (for plotting). 
The MATLAB-style tools are contained in the pyplot (plt) interface. 

This is really helpful for interactive plotting, because we can issue a command and see the result immediately. But, it is not suitable for more complicated cases. For these cases, we have another interface called **Object-Oriented** interface, described later.

## Object-Oriented Interface

### When to Use Object-Oriented Style:
- For **more complicated situations**
- When you want **more control over your figure**
- For creating multiple subplots or complex layouts

### Core Objects:

#### Figure Object:
- **Top-level container** for all plot elements
- Think of it as a **box-like container**
- Contains one or more **Axes** objects

#### Axes Object:
- Represents an **individual plot**
- Contains smaller objects such as:
  - **Axis** (x and y axes)
  - **Tick marks**
  - **Lines** (plot data)
  - **Legends**
  - **Title** and **text boxes**

For all Matplotlib plots, we start by creating a figure and an axes.
In their simplest form, a figure and axes can be created as follows:

````python
# First create a grid of plots
# ax will be an array of two Axes objects
# Create figure with 2 subplots stacked vertically
fig, ax = plt.subplots(2, 1)

# Call plot() method on the appropriate object
# Plot on first axes
ax[0].plot(x, np.sin(x)) # ax[0] = first subplot

# Plot on second axes
ax[1].plot(x, np.cos(x)) # ax[1] = second subplot
````

### Objects and Reference


The main idea with the **Object Oriented API** is to have objects that one can apply functions and actions on. The real advantage of this approach becomes apparent when more than one figure is created or when a figure contains more than one 
subplot.


We create a reference to the figure instance in the **fig** variable. Then, we ceate a new axis instance **axes** using the 
**add_axes** method in the Figure class instance fig as follows:-


````python
# Create figure and axes objects
fig = plt.figure() # Create figure
ax = plt.axes() # Create axes in the figure
````

## Understanding Figure and Axes Objects

### The Figure:
- An instance of the class `plt.Figure`
- Think of it as a **single container** that holds everything
- Contains all objects representing:
  - **Axes** (plot areas)
  - **Graphics** (lines, shapes, etc.)
  - **Text** (titles, labels)
  - **Labels** (axis labels, legends)

### The Axes:
- An instance of the class `plt.Axes`
- What we actually **see** when we look at a plot
- A **bounding box** with ticks and labels
- Will eventually contain the **plot elements** that make up our visualization

### Relationship:
- **Figure** = The entire canvas/container
- **Axes** = Individual plot area within that canvas
- One Figure can contain multiple Axes (subplots)

### Plotting with Axes:
- Once we have created axes, use `ax.plot()` function to plot data
- More explicit than pyplot's `plt.plot()`
- Gives direct control over which axes to plot on

### Example: Simple Sinusoid
Let's start with plotting a simple sine wave:

````python
# Plot using axes object
fig = plt.figure() # Create figure
ax = plt.axes() # Create axes in the figure

# 1000 points from 0 to 10
x = np.linspace(0, 10, 1000)
# Plot sine on axes object
ax.plot(x, np.sin(x));
# Plot cosine on axes
ax.plot(x, np.cos(x));
````

### Adjusting the Plot: Line Colors and Styles

The first adjustment you might wish to make to a plot is to control the line colors and styles.
The ``plt.plot()`` function takes additional arguments that can be used to specify these.
To adjust the color, you can use the ``color`` keyword, which accepts a string argument representing virtually any imaginable color.
The color can be specified in a variety of ways:

````python
# Different ways to specify colors in Matplotlib
plt.plot(x, np.sin(x - 0), color='blue')        # specify color by name
plt.plot(x, np.sin(x - 1), color='g')           # short color code (rgbcmyk)
plt.plot(x, np.sin(x - 2), color='0.75')        # Grayscale between 0 (black) and 1 (white)
plt.plot(x, np.sin(x - 3), color='#FFDD44')     # Hex code (RRGGBB from 00 to FF)
plt.plot(x, np.sin(x - 4), color=(1.0, 0.2, 0.3)) # RGB tuple, values 0 to 1
plt.plot(x, np.sin(x - 5), color='chartreuse'); # all HTML color names supported
plt.plot(x, np.sin(x - 300), color="#df0b8e"); # all HTML color names supported
````

If no color is specified, Matplotlib will automatically cycle through a set of default colors for multiple lines.

Similarly, the line style can be adjusted using the ``linestyle`` keyword:

````python
# Different line styles in Matplotlib
plt.plot(x, x + 0, linestyle='solid') # Solid line
plt.plot(x, x + 1, linestyle='dashed') # Dashed line
plt.plot(x, x + 2, linestyle='dashdot') # Dash-dot line
plt.plot(x, x + 3, linestyle='dotted'); # Dotted line

# For short, you can use the following codes:
plt.plot(x, x + 4, linestyle='-')  # solid
plt.plot(x, x + 5, linestyle='--') # dashed
plt.plot(x, x + 6, linestyle='-.') # dashdot
plt.plot(x, x + 7, linestyle=':');  # dotted
````

If you would like to be extremely terse, these ``linestyle`` and ``color`` codes can be combined into a single non-keyword argument to the ``plt.plot()`` function:

````python
# We can also specify color and line style together
plt.plot(x, x + 0, '-g')  # Solid green line
plt.plot(x, x + 1, '--c') # Dashed cyan line
plt.plot(x, x + 2, '-.k') # Dash-dot black line
plt.plot(x, x + 3, ':r');  # Dotted red line
````

### Adjusting the Plot: Axes Limits

Matplotlib does a decent job of choosing default axes limits for your plot, but sometimes it's nice to have finer control.
The most basic way to adjust axis limits is to use the ``plt.xlim()`` and ``plt.ylim()`` methods:

````python
# Set axis limits
plt.plot(x, np.sin(x)) # Plot sine function

plt.xlim(-1, 11) # Set x-axis from -1 to 11
plt.ylim(-1.5, 1.5) # Set y-axis from -1.5 to 1.5
````

If for some reason you'd like either axis to be displayed in reverse, you can simply reverse the order of the arguments:

````python
# Reverse axis order
plt.plot(x, np.sin(x)) # Plot sine function

plt.xlim(10, 0) # x-axis from 10 to 0 (reversed)
plt.ylim(1.2, -1.2) # y-axis from 1.2 to -1.2 (reversed)
````

A useful related method is ``plt.axis()`` (note here the potential confusion between *axes* with an *e*, and *axis* with an *i*).
The ``plt.axis()`` method allows you to set the ``x`` and ``y`` limits with a single call, by passing a list which specifies **``[xmin, xmax, ymin, ymax]``**:

````python
# Set both axes at once
plt.plot(x, np.sin(x)) # Plot sine function
# Set x and y axis limits together
# [xmin, xmax, ymin, ymax]
plt.axis([-1, 11, -1.5, 1.5])
````

The ``plt.axis()`` method goes even beyond this, allowing you to do things like automatically tighten the bounds around the current plot:

````python
# Auto-tight axis
plt.plot(x, np.sin(x)) # Plot sine function
plt.axis('tight') # Fit axes tightly around data
````

It allows even higher-level specifications, such as ensuring an equal aspect ratio so that on your screen, one unit in ``x`` is equal to one unit in ``y``:

````python
# Equal aspect ratio
plt.plot(x, np.sin(x)) # Plot sine function
# Make x and y units same size
plt.axis('equal') 
````

<font color=#FF0000 size='5'>**2. Seaborn: statistical data visualization**</font>

Seaborn is a library for making statistical graphics in Python. It builds on top of  matplotlib and integrates closely with pandas data structures.

Seaborn helps you explore and understand your data. ### Comparison of Seaborn and Matplotlib

Seaborn uses Matplotlib to draw plots. Many actions can be accomplished with only Seaborn functions. Further customization options might require using Matplotlib directly. Seaborn is not an alternative to Matplotlib. We can think of it as a complement to Matplotlib. As it is built on top of Matplotlib, we can call Matplotlib functions directly for creating simple plots.


It is summarized that 


“If Matplotlib tries to make easy things easy and hard things possible, Seaborn tries to make a well-defined set of hard things easy too in a well-defined way.” 

Seaborn helps to resolve two major problems faced by Matplotlib. The problems are:-

1.	Working with Matplotlib’s default parameters

2.	Working with data frames

- relational plots, 
- categorical plots, 
- distribution plots,

### Colour palette


Colour plays very important role in data visualization. Colour adds various dimensions to a plot when used effectively. 
A palette means a flat surface on which a painter mixes paints.


Seaborn provides a function called `color_palette()`. It can be used to give colours to plots and adding aesthetic value to it. It return a list of colors defining a color palette.


There are several readily available Seaborn palettes. These are: 
- Deep
- Muted
- Bright
- Pastel
- Dark
- Colorblind

Besides these we can also create new palettes.

There is another function `seaborn.palplot()` which deals with color palettes. This function plots the color palette as a horizontal array.


**Qualitative Color Palettes**

Qualitative or categorical palettes are best suitable to plot the categorical data as follows:-

````python
# Return a list of colors or continuous colormap defining a palette.
current_palette1 = sns.color_palette()
# Plot the values in a color palette as a horizontal array
sns.palplot(current_palette1)

plt.show()
````

We can see the desired number of colors by passing a value to the n_colors parameter. Here, the **`palplot()`** function is used to plot the array of colors horizontally.

----

**Sequential Color Palettes**

Sequential plots are suitable to express the distribution of data ranging from relative lower values to higher values within a range. Appending an additional character **"s"** to the color passed to the color parameter will plot the Sequential plot. 

We need to append 's' to the parameter like 'Greens' as follows:

````python
current_palette2 = sns.color_palette()

sns.palplot( sns.color_palette("Reds"))

plt.show()
````

**Diverging Color Palette**

Diverging palettes use two different colors. Each color represents variation in the value ranging from a common point in either direction. 

We assume plotting the data ranging from -1 to 1. The values from -1 to 0 takes one color and 0 to +1 takes another color. 

By default, the values are centered from zero. We can control it with parameter center by passing a value as follows:-


````python
current_palette3 = sns.color_palette()

sns.palplot( sns.color_palette("BrBG", 7))

plt.show()
````

#### Aggregation and Representing Uncertainty

### Complex Datasets:
- **More complex datasets** often have multiple measurements for the same `x` value
- Example: Multiple temperature readings for the same day
- Need to handle overlapping data points effectively

### Seaborn's Default Behavior:
When multiple measurements exist for the same `x` value, seaborn automatically:

1. **Aggregates** the multiple measurements at each `x` value
2. **Plots the mean** of those measurements
3. **Shows 95% confidence interval** around the mean

### What This Shows:
- **Central tendency**: The mean gives the typical value
- **Uncertainty**: The confidence interval shows the range of likely values
- **Data reliability**: Wider intervals suggest more uncertainty

### Benefits of This Approach:
- **Reduces visual clutter** from overlapping points
- **Shows both trend and uncertainty** in one visualization
- **Statistically meaningful** representation of the data
- **Easy to interpret** patterns and reliability

````python
# Load brain imaging data
fmri = sns.load_dataset("fmri")  # Functional MRI data
print("FMRI data first 5 rows:")
print(fmri.head())  # Show first 5 rows
````

# Boxplot

## What is a Boxplot?
- A **box plot (or box-and-whisker plot)** shows the distribution of quantitative data
- **Facilitates comparisons** between variables or across levels of a categorical variable
- Provides a **quick summary** of key distribution statistics

## What Boxplots Show:
### Quick Information:
- **Median** (center line)
- **Quartiles** (box boundaries)
- **Outliers** (individual points)

### What They Hide:
- **Individual data points** (except outliers)
- **Exact distribution shape**
- **Sample size**

## Anatomy of a Boxplot:

### The Box:
- **Outer bounds** represent the **first and third quartiles** (25th and 75th percentiles)
- **Line inside the box** represents the **median** (50th percentile)

### The Whiskers:
- **Extend to show the rest of the distribution**
- **Exclude points determined to be "outliers"**
- Length determined by **inter-quartile range (IQR)**

### The Outliers:
- **Markers outside the box**
- Represent **data points outside** the 25th and 75th percentiles
- Points that fall beyond the whisker boundaries

## Creating Boxplots:
- Use `boxplot(X)` to display **summary statistics** for a variable's distribution
- Efficient way to compare distributions across groups

````python
# Basic box plot
plt.figure()

# Show flipper length distribution by species
sns.boxplot(data=penguinsClean,
           x="species",             # Categories on x-axis
           y="flipper_length_mm")   # Values on y-axis
````

# Violin Plot

## What is a Violin Plot?
- A **combination of boxplot and kernel density estimate**
- Shows the **distribution of quantitative data** across levels of categorical variables
- Allows for **comparison of distributions** between groups

## Violin Plot vs Boxplot:

### Similarities:
- **Similar role** as box and whisker plots
- **Compares distributions** across categorical variables
- Shows **key statistical summaries**

### Key Differences:
- **Boxplot**: All components correspond to **actual datapoints**
- **Violin plot**: **Shows distribution shape** more clearly than boxplots

## Advantages:
- **Effective and attractive** way to show multiple distributions at once
- **Reveals distribution shape** (skewness, multiple peaks, etc.)
- **Combines summary statistics** with distribution visualization
- **More informative** than simple boxplots

## When to Use:
- **Large enough sample sizes** for reliable density estimation
- When **distribution shape matters** for your analysis
- For **comparing multiple groups** simultaneously

# Swarm Plot

## What is a Swarm Plot?
- A **categorical scatterplot** with points adjusted to be **non-overlapping**
- Shows **individual data points** while avoiding visual clutter
- Sometimes called a **"beeswarm"** plot

## How It Works:
- **Similar to `stripplot()`** but with a key difference
- **Points are adjusted** (only along the categorical axis) so they don't overlap
- Creates a **spread-out pattern** that resembles a swarm of bees

## Advantages:
- **Better representation** of the distribution of values than overlapping points
- **Shows every individual observation**
- **Reveals data density** patterns clearly
- **No information loss** - every data point is visible

## Limitations:
- **Does not scale well** to large numbers of observations
- Can become **cluttered** with too many points
- **Performance issues** with very large datasets

## Best Use Cases:

### Standalone:
- **Can be drawn on its own** for simple data exploration
- Good for **small to medium datasets**

### As a Complement:
- **Good complement** to box or violin plots
- **Shows all observations** along with distribution summary
- **Combines individual data** with statistical summaries
- Provides **complete picture** of the data

````python
# Show body mass distribution by species using swarm plot
sns.swarmplot(data=penguinsClean,
             x='species',           # Categories
             y='body_mass_g')       # Values
````

````python
# Complex time series plot
import seaborn as sns

# Includes number of passenger information for each flight by year
flights = sns.load_dataset("flights")  # Load flight data

# Create multi-panel plot
g = sns.relplot(data=flights,
               x="month",               # x-axis
               y="passengers",          # y-axis
               col="year",              # Separate plot per year
               hue="year",              # Color by year
               kind="line",             # Line plot
               linewidth=4,             # Thick lines
               zorder=5,                # Draw on top
               col_wrap=3,              # 3 columns per row
               height=2,                # Plot height
               aspect=1.5,              # Width/height ratio
               legend=False)            # No legend

# Customize each subplot
for year, ax in g.axes_dict.items():
    # Add year as text
    ax.text(.8, .85, year,              # Position and text
           transform=ax.transAxes,      # Use axes coordinates
           fontweight="bold")           # Bold font
    
    # Add background lines for all years
    sns.lineplot(data=flights,
                x="month",              # x-axis
                y="passengers",         # y-axis
                units="year",           # Group by year
                estimator=None,         # No averaging
                color=".7",             # Gray color
                linewidth=1,            # Thin lines
                ax=ax)                  # Current axes

# Reduce x-axis tick frequency
ax.set_xticks(ax.get_xticks()[::2])     # Every other tick

# Tweak the supporting aspects of the plot
g.set_titles("")                        # Remove titles
g.set_axis_labels("", "Passengers")     # Set axis labels
g.tight_layout()                        # Adjust spacing
plt.show()
````

<font color='maroon' size='5'>Building structured multi-plot grids</font> 
----

- When exploring multi-dimensional data, a useful approach is to draw multiple instances of the same plot on different subsets of your dataset. This technique is sometimes called either “lattice” or “trellis” plotting, and it is related to the idea of “small multiples”. It allows a viewer to quickly extract a large amount of information about a complex dataset. 

    - Matplotlib offers good support for making figures with multiple axes; seaborn builds on top of this to directly link the structure of the plot to the structure of your dataset.


- **Conditional small multiples**
    - The `FacetGrid` class is useful when you want to visualize the distribution of a variable or the relationship between multiple variables separately within subsets of your dataset. 
    - A FacetGrid can be drawn with up to three dimensions: row, col, and hue. 
    - The first two have obvious correspondence with the resulting array of axes; think of the hue variable as a third dimension along a depth axis, where different levels are plotted with different colors.
- Each of `relplot()`, `displot()`, `catplot()`, and `lmplot()`, use this object internally, and they return the object when they are finished so that it can be used for further tweaking.

- The class is used by initializing a `FacetGrid` object with a dataframe and the names of the variables that will form the row, column, or hue dimensions of the grid. These variables should be categorical or discrete, and then the data at each level of the variable will be used for a facet along that axis. 

# FacetGrid

## What is FacetGrid?
- **Multi-plot grid** for plotting conditional relationships
- Creates **multiple subplots** based on categorical variables

## When to Use FacetGrid:
- When you want to **visualize distribution** of a variable within subsets
- When you want to see **relationships between variables** separately for different groups
- For **comparing patterns** across categories

## Three Dimensions Available:

### 1. Row Dimension (`row`):
- Creates **separate rows** of plots
- Each row represents a different level of the row variable

### 2. Column Dimension (`col`):
- Creates **separate columns** of plots  
- Each column represents a different level of the column variable

### 3. Hue Dimension (`hue`):
- **Third dimension along depth axis**
- Different levels plotted with **different colors**
- Adds color coding within each subplot

## How to Use FacetGrid:

### Initialization:
1. **Initialize FacetGrid object** with a dataframe
2. **Specify variable names** for `row`, `column`, or `hue` dimensions
3. **Variables should be categorical or discrete**

### Data Organization:
- **Data at each level** of the variable creates a facet along that axis
- Each unique combination gets its own subplot
- Enables **systematic comparison** across groups

## Key Benefits:
- **Systematic exploration** of conditional relationships
- **Easy comparison** across multiple categories
- **Reduces cognitive load** by organizing complex data

````python
# FacetGrid example
# Create grid of plots by island and sex
# Color by species
g = sns.FacetGrid(penguinsClean,
                 col="island",          # Columns by island
                 row="sex",             # Rows by sex
                 hue="species")         # Color by species

# Map histogram to each subplot
g.map(sns.histplot,                     # Plot type
     "flipper_length_mm",               # Variable
     alpha=.7)                          # Transparency

g.add_legend()                          # Add legend
````

## Other Python Graphics Libraries

Although Matplotlib is the most prominent Python visualization library, there are other more modern tools that are worth exploring as well.
I'll mention a few of them briefly here:

- [Bokeh](http://bokeh.pydata.org) is a JavaScript visualization library with a Python frontend that creates highly interactive visualizations capable of handling very large and/or streaming datasets. The Python front-end outputs a JSON data structure that can be interpreted by the Bokeh JS engine.
- [Plotly](http://plot.ly) is the eponymous open source product of the Plotly company, and is similar in spirit to Bokeh. Because Plotly is the main product of a startup, it is receiving a high level of development effort. Use of the library is entirely free.
- [Vispy](http://vispy.org/) is an actively developed project focused on dynamic visualizations of very large datasets. Because it is built to target OpenGL and make use of efficient graphics processors in your computer, it is able to render some quite large and stunning visualizations.
- [Vega](https://vega.github.io/) and [Vega-Lite](https://vega.github.io/vega-lite) are declarative graphics representations, and are the product of years of research into the fundamental language of data visualization. The reference rendering implementation is JavaScript, but the API is language agnostic. There is a Python API under development in the [Altair](https://altair-viz.github.io/) package. Though as of summer 2016 it's not yet fully mature, I'm quite excited for the possibilities of this project to provide a common reference point for visualization in Python and other languages.

The visualization space in the Python community is very dynamic, and I fully expect this list to be out of date as soon as it is published.
Keep an eye out for what's coming in the future!
