#### PieChart - Part of A Whole

A Pie Chart is a circle divided into sectors that each represent a proportion of the whole. It is one of the most common viz type, but also probably the most criticized.

#### Matplotlib

```python
# library
import matplotlib.pyplot as plt
import numpy as np

# Make a random dataset:
popularity = [56, 39, 34, 34, 29]

# Create a pieplot
plt.pie(popularity)
plt.show()
```

#### Add labels

Labels are added thanks to the labels parameter that expect an array of strings. You also problably want to play with the labeldistance parameter that controls the distance between the circle edge and the label itself. (1= exactly on it, use something bigger to have the label further.)

```python
# library
import matplotlib.pyplot as plt
import numpy as np

# Make a random dataset:
languages =['Python', 'SQL', 'Java', 'C++', 'JavaScript']
popularity = [35, 15, 18, 25, 7]

# Create a pieplot
plt.pie(popularity, labels = languages, labeldistance = 1.15, autopct='%.1f%%')
plt.show()
```

#### Customize wedges

It is pretty common to customize wedges to get some blank spaces around them. This is doable thanks to the wedgeprops parameter. This parameter is an object that accept 2 properties: linewidth and edgecolor



```python
# library
import matplotlib.pyplot as plt
import numpy as np

# Make a random dataset:
languages =['Python', 'SQL', 'Java', 'C++', 'JavaScript']
popularity = [35, 15, 18, 25, 7]

# Create a pieplot
plt.pie(popularity, labels = languages, labeldistance = 1.15, autopct='%.1f%%', wedgeprops = { 'linewidth' : 3, 'edgecolor' : 'white'})
plt.show()
```

#### Styling the Pie Chart

You can use the wedgeprops and textprops arguments to style the wedges and texts, respectively.

```python
# library
import matplotlib.pyplot as plt
import numpy as np

# Make a random dataset:
languages =['Python', 'SQL', 'Java', 'C++', 'JavaScript']
popularity = [35, 15, 18, 25, 7]

# # Capture each of the return elements.
patches, texts, pcts = plt.pie(popularity, labels = languages, 
                               labeldistance = 1.15,
                               autopct='%.1f%%', 
                               wedgeprops = { 'linewidth' : 3, 'edgecolor' : 'white'})

# Style just the percent values.
plt.setp(pcts, color='white', fontweight='bold')
plt.show()
```

```python
# library
import matplotlib.pyplot as plt
import numpy as np

# Make a random dataset:
languages =['Python', 'SQL', 'Java', 'C++', 'JavaScript']
popularity = [35, 15, 18, 25, 7]

# # Capture each of the return elements.
patches, texts, pcts = plt.pie(popularity, labels = languages, 
                               labeldistance = 1.15,
                               autopct='%.1f%%', 
                               wedgeprops = { 'linewidth' : 3, 'edgecolor' : 'white'})


# For each wedge, set the corresponding text label color to the wedge's
# face color.
for i, patch in enumerate(patches):
  texts[i].set_color(patch.get_facecolor())

# Style just the percent values.
plt.setp(pcts, color='white', fontweight='bold')
plt.show()
```

```python
# library
import matplotlib.pyplot as plt
import numpy as np

# Make a random dataset:
languages =['Python', 'SQL', 'Java', 'C++', 'JavaScript']
popularity = [35, 15, 18, 25, 7]

# Get four different grey colors.
cmap = plt.get_cmap('Greys')
colors = list(cmap(np.linspace(0.45, 0.85, len(languages))))
# Swap in a bright blue for the Lacrosse color.
colors[0] = 'dodgerblue'

# # Capture each of the return elements.
patches, texts, pcts = plt.pie(popularity, labels = languages, 
                               labeldistance = 1.15,
                               autopct='%.1f%%',
                               colors = colors,
                               wedgeprops = { 'linewidth' : 3, 'edgecolor' : 'white'},
                               explode = (0.1, 0, 0, 0, 0))


# For each wedge, set the corresponding text label color to the wedge's
# face color.
for i, patch in enumerate(patches):
  texts[i].set_color(patch.get_facecolor())

# Style just the percent values.
plt.setp(pcts, color='white', fontweight='bold')
plt.show()
```
