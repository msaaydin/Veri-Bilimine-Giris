### Scatterplot - Correlation

A scatter plot displays the relationship between 2 numeric variables. Each data point is represented as a circle. Several tools allow to build one in python, this section provides code samples for Seaborn, Matplotlib

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Read Data
flights = sns.load_dataset("flights")
flights.head()
```

#### Matplotlib

```python
## We want to find correlation between Month and Number of Passengers

data = flights.query("year == 1951") # Return lines that year is '1951'
plt.scatter(data['month'], data['passengers']) # Draw Scatter Plot
plt.xlabel("Month")
plt.ylabel("Number of Passengers")
plt.show()
```

#### Seaborn

```python
data = flights.query("year == 1951") # Return lines that year is '1951'

sns.scatterplot(data = data, x = 'month', y = 'passengers') # Draw Scatter Plot
```

### Scatter Plot with Grouping Data

#### Matplotlib

```python
import matplotlib.pyplot as plt
import seaborn as sns
iris = sns.load_dataset("iris")

# Return each species in iris data
species = list(iris.groupby("species").groups.keys())
# 
colors = {'setosa': 'b', 'versicolor' : 'orange', 'virginica' : 'g'}

for name in species:
    data = iris.query("species == @name")
    plt.scatter(data['sepal_length'], data['petal_width'], color=colors[name], label=name)

plt.xlabel("Sepal Length")
plt.ylabel("Petal Width")
plt.legend()
```

#### Seaborn

```python
import seaborn as sns
iris = sns.load_dataset("iris")

species = list(iris.groupby("species").groups.keys())
sns.scatterplot(data = iris, x = 'sepal_length', y = 'petal_width', hue='species')
```
